```python
df_excel=pd.read_excel("Country-Code.xlsx")
df.head(2)
```

```python
df_n=pd.merge(df,df_excel,on="Country Code",how='inner')
df_n.head(2)
```

```python
# To check data types
df_n.dtypes
```

```python
df_n.Country.value_counts()
```

```python
v=df_n["Country"].value_counts()
County=v.index.tolist()
Values=v.values.tolist()
print(County,Values)
```

```python
# Pie chart
plt.pie(Values[:3],labels=County[:3],autopct="%1.2f%%",shadow=True,radius=1)
plt.legend()
plt.title("Country-wise share of Zomato users",loc="right")
plt.grid(True,linestyle="--",linewidth="0.3",color="gray")
plt.show()
```

Observations:Maximum market share of Zomato is in India followed by United States.

```python
df_n.columns
```

```python
ratings=df_n.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:"Rating count"})
ratings.head(10)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating count",data=ratings,palette="viridis")
```

Observatin:  
1.Not rated count is very high  
2.maximum rating is between 2.5 to 3.9  

```python
sns.countplot(x="Rating color",data=ratings,palette=["blue","red","orange","yellow","green","green"])
```
