# Descriptive Statistics

# Measue of Central Tendency
# 1.Mean 
# 2.Median
# 3.Mode

```python
import numpy as np
import pandas as pd
import seaborn as sns
import statistics as s
```

```python
np.random.seed(42)
data=np.random.randint(10,50,40)
```

```python
data
```

```python
print(np.mean(data))
print(np.median(data))
```

```python
#Finiding central tendency usinf statistics module of python
import statistics as s
print(s.mean(data))
print(s.median(data))
print(s.mode(data))
```

```python
data1=[12,10,15,23,41,32,30,25,29,35,39,12,10,42,45,15,29,21,15,30,100,120]
sns.boxplot(data1,orient="h")
```

# Five number Summary
# Min,1st Quartile(Q1),median,3rd Quartile(Q3),Max

```python

print("Minimum =",min(data))
print("Maximum =",max(data))
Q1=np.percentile(data,[25])
Q3=np.percentile(data,[75])
print("Q1 =",Q1)
print("Q3 =",Q3)
IQR=Q3-Q1
print("Lower Fence =",Q1-1.5*(IQR))
print("Upper Fence =",Q3+1.5*(IQR))
```

# Measure of Dispersion
# 1.Variance
# 2.Standard Deviation

```python
def variance(data,dof=1):
    n=len(data)
    mean=(sum(data)/n)
    deviation=[(x-mean)**2 for x in data]
    variance=sum(deviation)/(n-dof)
    return variance
print(variance(data))
print(variance(data,0))
```

```python
print(np.var(data))
print(np.var(data))
```

```python
print(s.pvariance(data))
```

# Differnet Distributions

# Normal Distribution

```python
d1=np.random.normal(0.5,0.2,1000)*100
```

```python
sns.histplot(d1,kde=True)
```

# Log normal Distribution

```python
d2=np.random.lognormal(3.0,1.0,1000)
sns.histplot(d2,kde=True)
```

```python
# log of log normal distribution gives normal distribution
sns.histplot(np.log(d2),kde=True)
```

# Check whether a distribution is normal or not

```python
#we use Q-Q plot
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
import numpy as np
import seaborn as sns
def plot_data(sample):
    fig, axes = plt.subplots(1, 2, figsize=(10, 4))
    sns.histplot(sample,ax=axes[0])
    stat.probplot(sample,dist='norm',plot=pylab)
    plt.show()
    
```
